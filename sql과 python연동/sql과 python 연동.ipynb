```python
world = mysql.connector.connect(host='localhost',user='inab1822',password='neinei955',
                               db='world',charset='utf8')
cur = world.cursor()
sql = 'select name, continent, region, population, gnp, gnpold, governmentform from country;'
cur.execute(sql)
rows = cur.fetchall()
world.close()
world_df = pd.DataFrame(rows)
world_df.columns = ['name','continent','region','population',
                        'gnp','gnpold','governmentform']
```

```python
world_df
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
# world_df 숫자데이터로변환
world_df.info()
```

```python
world_df['gnp'] = world_df['gnp'].astype(float)
world_df['gnpold'] = world_df['gnpold'].astype(float)
```

```python
world_df.info() # 적용완료
```

```python
# gnp를 계산해 각 나라마다 순위를 매겨주었다
# 같은 gnp를 가지면 동순위를 매긴다.
world_df['rank'] = world_df['gnp'].rank(method='dense',ascending=False).astype(int)
```

```python
# 만약 gnp의 값이 결측치 이거나 0.0 이면 rank컬럼의 값에 np.nan을 넣겠다.
world_df.loc[world_df['gnp']==0.0,'rank'] = np.nan
```

```python
world_df.sort_values('rank',ascending=True)
```

```python
# 대한민국은 전체 gnp랭킹 16위이다.
world_df[world_df['name']=='South Korea']
```

```python
# gnp 랭킹 20위 까지만 출력해보자.
world_df_gnp_top20 = world_df.sort_values('rank',ascending=True).head(20)
```

```python
world_df_gnp_top20
```

```python
# 랭킹 20위 안 gnp순위
plt.figure(figsize=(12,8))
plt.xticks(rotation=45)
sns.barplot(data=world_df_gnp_top20, x='name',y='gnp')
```

```python
# gnp순위 top100안에서 아시아대륙안의 gnp순위
world_df_gnp_top100 = world_df.sort_values('rank',ascending=True).head(100)
plt.figure(figsize=(12,8))
plt.xticks(rotation=45)
sns.barplot(data=world_df_gnp_top100[world_df_gnp_top100['continent']=='Asia'],
            x='name',y='gnp')
```
